```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('/content/sample_data/california_housing_test.csv')
df_test = pd.read_csv('/content/sample_data/california_housing_train.csv')

x = df_train.values
x = x[:500,:5]
y = df_test.values
y = y[:500,:5]
print(y.shape)
print(x.shape)

plt.scatter(x,y)
```

```python
#normalisation means mean should be zero and standard deviation should be one
x2 = (x - x.mean())/ x.std()
#y2 = (y - y.mean())/ y.std()
plt.scatter(x,y, label = "raw data")
plt.scatter(x2, y, color = 'orange', label = "Normalised data")
plt.legend()
plt.show()
#now new data is having zero mean , we can se below in graph
```
